```python
def printProp(prop_name, output_file=None):
    if output_file == None:
        try:
            sp = security_properties[prop_name]
            print("//# of Generic Signals to Replace = {}".format(sp[1]))
            print(sp[0])
            print("")
        except KeyError:
            print("Error: \"{}\" is not a valid security property name".format(prop_name))
    else:
        try:
            sp = security_properties[prop_name]
            print("//# of Generic Signals to Replace = {}".format(sp[1]), file=output_file)
            print(sp[0], file=output_file)
            print("", file=output_file)
        except KeyError:
            print("Error: \"{}\" is not a valid security property name".format(prop_name))

def printAllProps():
    for key in security_properties.keys():
        printProp(key)

def useProp(prop_name, design_signals=list(), prop_label=""):
    try:
        sp = security_properties[prop_name]
        
        if len(design_signals) != sp[1]:
            print("Error: # of design_signals ({}) does not match # of generic signals to replace ({})"\
                  .format( len(design_signals), sp[1]))
            printProp(prop_name)
            return None
        
        if sp[1] == 0:
            return sp[0]
        
        #Replace generic signals with design signals
        split_prop = str.split(sp[0], "`")

        sig_in_split_index = 1
        for signal in design_signals:
            split_prop[sig_in_split_index] = signal
            sig_in_split_index += 2
        new_sp = "".join(split_prop)
        
        #Replace generic property label
        new_sp = str.split(new_sp, "GENERIC")
        return new_sp[0]+prop_label+new_sp[1]
    except KeyError:
        print("Error: \"{}\" is not a valid security property name".format(prop_name))

def genPropLabelFromSig(name_A, name_B):
    label = ""
    for i in range(min(len(name_A), len(name_B))):
        if name_A[i] == name_B[i]:
            label += name_A[i]
    if label[-1] == "_":
        label = label[:-1]
    return label

def genSpecificProps(prop_name, replacements, as_file, label_id_index = None):
    for i, sig_pair in enumerate(replacements):
        label_id = None
        if label_id_index == None:
            label_id = genPropLabelFromSig(sig_pair[0], sig_pair[1])
        elif type(label_id_index)==int:
            label_id = sig_pair[label_id_index]
        print(useProp(prop_name,sig_pair, label_id), file=as_file)
        print("",file=as_file)

def genDivider(as_file):
    print("////////////////////////////////////////",file=as_file)
    print("////////////////////////////////////////",file=as_file)
    print("",file=as_file)
```
